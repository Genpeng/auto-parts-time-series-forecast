## 1 数据预处理

```python
# 加载数据
data_path = "../data/engine_oil.txt"
df = pd.read_csv(data_path, sep='\t', encoding='utf-8')
```

```python
df = df.rename(columns={'物料编码': 'part_id', '物料描述': 'part_name', 
                        '订货数': 'order_num', '缺件数': 'out_of_stock_num', 
                        '受理数': 'delivery_num', '审核日期': 'date', '审核时间': 'time'})

# 将`part_id`的数据类型设为字符串，方便后面进行重采样 
df['part_id'] = df['part_id'].astype('str')

# 重置索引
df['date'] = pd.to_datetime(df['date'], format="%Y-%m-%d")
df.set_index('date', inplace=True)
```

```python
# 去除2016年10月份的数据
df = df[3:]
```

```python
# 重采样
df_month = df.resample('M').sum()
```

```python
train_month = df_month[:-4]
test_month = df_month[-4:]
```

```python
fig, ax = plt.subplots()
ax.plot(df_month.index, df_month.order_num.values, 'bo-')
ax.set_xlabel('month', fontsize=16)
ax.set_ylabel('sales', fontsize=16)
ax.set_title('机油每月订货量', fontsize=20)
ax.grid()

# TODO: 改变刻度字体大小不够优雅
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

## 2 Holt-Winters method

### 2.1 按照天

```python
model = ExponentialSmoothing(train_day.order_num.values, seasonal_periods=7, trend='add', seasonal='add')
model_fit = model.fit()
pred = model_fit.forecast(120)
```

```python
plt.figure(figsize=(15, 8))
plt.plot(train_day.index, train_day.order_num.values, label='Train')
plt.plot(test_day.index, test_day.order_num.values, label='Test')
plt.plot(test_day.index, pred, label='Holt-Winters method')
plt.title("Holt-Winters方法预测每天的机油订货数")
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
rmse = sqrt(mean_squared_error(test_day.order_num.values, pred))
print("The RMSE of 'Holt-Winters method' is:", rmse)
```

### 2.2 按照周

```python
model = ExponentialSmoothing(train_week.order_num.values, seasonal_periods=4, trend='add', seasonal='add')
model_fit = model.fit()
pred = model_fit.forecast(16)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(train_week.index, train_week.order_num.values, label='Train')
plt.plot(test_week.index, test_week.order_num.values, label='Test')
plt.plot(test_week.index, pred, label='Holt-Winters method')
plt.title("Holt-Winters method")
plt.legend(loc='best')
plt.show()
```

```python
rmse = sqrt(mean_squared_error(test_week.order_num.values, pred))
print("The RMSE of 'Holt-Winters method' is:", rmse)
```

## 3 ARIMA

### 3.1 按照天

```python
train, test = train_day.order_num.values, test_day.order_num.values
```

```python
history = list(train)
preds = list()
for i in range(len(test)):
    model = ARIMA(history, order=(7, 0, 0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    preds.append(yhat)
    obs = test[i]
    history.append(obs)
    print("expected = %f predicted = %f" % (obs, yhat))
error = sqrt(mean_squared_error(test, preds))
print("Test RMSE: %.3f" % error)
```

```python
fig, ax = plt.subplots()
ax.plot(test_day.index, test, 'bo-', label='test')
ax.plot(test_day.index, preds, 'ro-', label='pred')
ax.set_xlabel('month', fontsize=16)
ax.set_ylabel('days', fontsize=16)
ax.set_title('ARIMA拟合机油每天订货量', fontsize=20)
ax.legend(loc='upper left')
ax.grid()

# TODO: 改变刻度字体大小不够优雅
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

fig.savefig("../figs/ARIMA预测图.png")
plt.show()
```

### 按照月

```python
train, test = train_month.order_num.values, test_month.order_num.values
```

```python
history = list(train)
preds = list()
for i in range(len(test)):
    model = ARIMA(history, order=(2, 0, 0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    preds.append(yhat)
    obs = test[i]
    history.append(obs)
    print("expected = %f predicted = %f" % (obs, yhat))
error = sqrt(mean_squared_error(test, preds))
print("Test RMSE: %.3f" % error)
```

```python
fig, ax = plt.subplots()
ax.plot(test_month.index, test, 'bo-', label='test')
ax.plot(test_month.index, preds, 'ro-', label='pred')
ax.set_xlabel('month', fontsize=16)
ax.set_ylabel('sales', fontsize=16)
ax.set_title('ARIMA Rolling Forecast Line Plot', fontsize=20)
ax.legend(loc='upper left')
ax.grid()

# TODO: 改变刻度字体大小不够优雅
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

fig.savefig("../figs/ARIMA预测图.png")
plt.show()
```
