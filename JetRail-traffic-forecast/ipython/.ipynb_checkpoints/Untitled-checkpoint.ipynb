```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.api import ExponentialSmoothing, ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX

%matplotlib inline
```

```python
# 加载全部数据
# data_path = "../data/Train_SU63ISt.csv"
# df = pd.read_csv(data_path)
```

```python
# 加载部分数据（2012年8月到2013年12月）
data_path = "../data/Train_SU63ISt.csv"
df = pd.read_csv(data_path, nrows=11856)
```

```python
# 重置索引
df['Timestamp'] = pd.to_datetime(df.Datetime, format="%d-%m-%Y %H:%M")
df.set_index('Timestamp', inplace=True)
```

```python
# 删除冗余的列
df.drop(['Datetime'], axis=1, inplace=True)
```

```python
# 划分数据集
train_df = df[:10392]
test_df = df[10392:]
```

```python
series = df.Count.resample('D').mean()
series_train = train_df.Count.resample('D').mean()
series_test = test_df.Count.resample('D').mean()
```

```python
series_train.plot(figsize=(15, 8), title="Daily ridership", fontsize=14)
series_test.plot(figsize=(15, 8), title="Daily ridership", fontsize=14)
plt.show()
```

## 1 Naive method

```python
pred_value = np.asarray(series_train)[-1]
y_hat = np.array([pred_value] * len(series_test))
```

```python
plt.figure(figsize=(15, 8))
plt.plot(series_train, label='Train')
plt.plot(series_test, label='Test')
plt.plot(series_test.index, y_hat, label='Naive method')
plt.title("Naive method")
plt.legend(loc='best')
plt.show()
```

```python
rmse = sqrt(mean_squared_error(series_test.values, y_hat))
print("The RMSE of 'Naive method' is:", rmse)
```

## 2 Holt-Winters method

```python
model = ExponentialSmoothing(series_train.values, seasonal_periods=7, trend='add', seasonal='add')
model_fit = model.fit()
y_hat = model_fit.forecast(len(series_test))
```

```python
plt.figure(figsize=(15, 8))
plt.plot(series_train, label='Train')
plt.plot(series_test, label='Test')
plt.plot(series_test.index, y_hat, label='Holt-Winters method')
plt.title('Holt-Winters method')
plt.legend(loc='best')
plt.show()
```

```python
rmse = sqrt(mean_squared_error(series_test.values, y_hat))
print("The RMSE of 'Holt-Winters method' is:", rmse)
```

## 3 ARIMA

```python
model = SARIMAX(series_train, order=(2, 1, 4), seasonal_order=(0, 1, 1, 7))
model_fit = model.fit()
y_hat = model_fit.predict(start='2013-11-1', end='2013-12-31', dynamic=True)
```

```python
plt.figure(figsize=(15, 8))
plt.plot(series_train, label='Train')
plt.plot(series_test, label='Test')
plt.plot(series_test.index, y_hat, label='ARIMA')
plt.title('ARIMA')
plt.legend(loc='best')
plt.show()
```

```python
rmse = sqrt(mean_squared_error(series_test.values, y_hat))
print("The RMSE of ARIMA' is:", rmse)
```
